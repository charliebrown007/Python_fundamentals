- The layout is organized in rows and columns, which are represented by the first and second argument.
- The third argument represents the index of the current plot.

## Histograma

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

For simplicity we use NumPy to randomly generate an array with 250 values, where the values will concentrate around 170, and the standard deviation is 10

```python
x = np.random.normal(170, 10, 250)
print(x)
```

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

## Gr√°fico de torta

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

Etiquetas en un gr√°fico de torta

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

Selecci√≥n del √°ngulo de inicio

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

Separaci√≥n de porciones

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0.1, 0.1, 0.1]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

Sombras

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

Colores

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

Leyendas de gr√°fico

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
